```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data titik-titik 2D
points = np.array([
    [2, 3],
    [5, 4],
    [9, 6],
    [4, 7],
    [8, 1],
    [7, 2]
])
```

```python
# Fungsi untuk membangun KD-Tree dan partisi ruang
def kdtree(points, depth=0):
    n = len(points)
    if n <= 0:
        return None

    # Pilih dimensi pembagian (x atau y) bergantian
    k = points.shape[1]  # 2D jadi k=2
    axis = depth % k

    # Urutkan dan pilih median untuk pembagian
    sorted_points = points[points[:, axis].argsort()]
    median = n // 2

    # Buat node KD-Tree dan partisi secara rekursif
    return {
        'point': sorted_points[median],
        'left': kdtree(sorted_points[:median], depth + 1),
        'right': kdtree(sorted_points[median + 1:], depth + 1),
        'axis': axis
    }
```

```python
# Fungsi untuk menggambar partisi KD-Tree
def plot_kdtree(node, min_x, max_x, min_y, max_y, depth=0):
    if node is None:
        return
    
    x, y = node['point']
    axis = node['axis']

    # Tentukan batas partisi berdasarkan dimensi
    if axis == 0:  # Partisi di sepanjang sumbu-x
        plt.plot([x, x], [min_y, max_y], 'r--')  # Garis vertikal
        plot_kdtree(node['left'], min_x, x, min_y, max_y, depth + 1)
        plot_kdtree(node['right'], x, max_x, min_y, max_y, depth + 1)
    else:  # Partisi di sepanjang sumbu-y
        plt.plot([min_x, max_x], [y, y], 'b--')  # Garis horizontal
        plot_kdtree(node['left'], min_x, max_x, min_y, y, depth + 1)
        plot_kdtree(node['right'], min_x, max_x, y, max_y, depth + 1)

    # Plot titik data
    plt.plot(x, y, 'ko')
    plt.text(x + 0.1, y + 0.1, f'({x}, {y})', fontsize=10)
```

```python
# Membangun KD-Tree
kdtree_root = kdtree(points)

# Plot data dan partisi
plt.figure(figsize=(8, 8))
plt.xlim(0, 10)
plt.ylim(0, 10)
plot_kdtree(kdtree_root, min_x=0, max_x=10, min_y=0, max_y=10)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Partisi KD-Tree pada Data 2D")
plt.grid(True)
plt.show()
```
